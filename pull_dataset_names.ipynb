```python
#!/usr/bin/env python
import urllib.request
import json
import time

size = 1000
start_i = 0
all_results = []

while True:
    try:
        # Make the HTTP request.
        response = urllib.request.urlopen(f'http://catalog.data.gov/api/3/action/package_search?rows={size}&start={start_i}')

        if response.code != 200:
            break
            
        response_dict = json.loads(response.read())
        
        if response_dict['success'] is not True:
            break
        
        results = response_dict['result']['results']

        if len(results) == 0:
            break

        results = [r['title'].replace('\n', '') for r in results]

        all_results.extend(results)
        print(len(all_results))
        
        start_i += size
        time.sleep(1)
        
        """if len(all_results) > 1000:
            break"""
    except Exception as e:
        print(e)
        break
```

```python
print('Saving results...')
with open('data/dataset_names.txt', 'w', encoding="utf-8") as f:
    for l in all_results:
        try:
            f.write(l + '\n')
        except Exception as e:
            print(e)

print('Saved.')
```
